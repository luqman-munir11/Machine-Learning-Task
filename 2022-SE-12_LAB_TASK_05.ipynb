```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("taxi_trip_pricing.csv")
```

```python
df.describe()
```

```python
df[df['Trip_Distance_km'].isnull()]
```

```python
df['Trip_Distance_km'].fillna(df['Trip_Distance_km'].median(), inplace=True)
```

```python
df
```

```python
df[df['Trip_Distance_km'].isnull()].head(5)
```

```python
df['Passenger_Count'].fillna(df['Passenger_Count'].median(), inplace=True)
df['Base_Fare'].fillna(df['Base_Fare'].median(), inplace=True)
df['Per_Km_Rate'].fillna(df['Per_Km_Rate'].median(), inplace=True)
df['Per_Minute_Rate'].fillna(df['Per_Minute_Rate'].median(), inplace=True)
df['Trip_Duration_Minutes'].fillna(df['Trip_Duration_Minutes'].median(), inplace=True)
df['Trip_Price'].fillna(df['Trip_Price'].median(), inplace=True)
```

```python
df['Time_of_Day'].fillna(df['Time_of_Day'].mode(), inplace=True)
df['Day_of_Week'].fillna(df['Day_of_Week'].mode(), inplace=True)
df['Traffic_Conditions'].fillna(df['Traffic_Conditions'].mode(), inplace=True)
df['Weather'].fillna(df['Weather'].mode(), inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
df[['Trip_Distance_km', 'Passenger_Count', 'Base_Fare','Per_Km_Rate', 'Per_Minute_Rate', 'Trip_Duration_Minutes', 'Trip_Price']].corr()
```

```python
sns.scatterplot(X='Trip_Distance_km',y='Trip_Price',data=df)
```

```python
X=df[['Trip_Distance_km']]
y=df['Trip_Price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
np.sqrt(mse)
```

```python
# Scatterplot of actual data
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot the regression line
plt.plot(X, model.predict(X), color='red', label='Regression Line')

# Plot errors (residuals)
for i in range(len(X)):
    plt.plot([X.iloc[i, 0], X.iloc[i, 0]], [y.iloc[i], model.predict(X)[i]], color='green', linestyle='dotted')

# Labels and title
plt.xlabel('Trip_Distance_km')
plt.ylabel('Trip_Price')
plt.title('Regression Line and Errors')
plt.legend()
plt.show()
```
