```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("taxi_trip_pricing.csv")
```

```python

df['Trip_Distance_km'].fillna(df['Trip_Distance_km'].median(), inplace=True)
df['Passenger_Count'].fillna(df['Passenger_Count'].median(), inplace=True)
df['Base_Fare'].fillna(df['Base_Fare'].median(), inplace=True)
df['Per_Km_Rate'].fillna(df['Per_Km_Rate'].median(), inplace=True)
df['Per_Minute_Rate'].fillna(df['Per_Minute_Rate'].median(), inplace=True)
df['Trip_Duration_Minutes'].fillna(df['Trip_Duration_Minutes'].median(), inplace=True)
df['Trip_Price'].fillna(df['Trip_Price'].median(), inplace=True)
```

```python
df.describe()
```

```python
df['Time_of_Day'].fillna(df['Time_of_Day'].mode(), inplace=True)
df['Day_of_Week'].fillna(df['Day_of_Week'].mode(), inplace=True)
df['Traffic_Conditions'].fillna(df['Traffic_Conditions'].mode(), inplace=True)
df['Weather'].fillna(df['Weather'].mode(), inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
df[['Trip_Distance_km', 'Passenger_Count', 'Base_Fare','Per_Km_Rate', 'Per_Minute_Rate', 'Trip_Duration_Minutes', 'Trip_Price']].corr()
```

```python
sns.scatterplot(x='Trip_Distance_km',y='Trip_Price',data=df)
```

```python
X = df[["Trip_Distance_km", "Per_Km_Rate", "Trip_Duration_Minutes"]]
y = df['Trip_Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[['Trip_Distance_km', 'Per_Km_Rate', 'Trip_Duration_Minutes']])
X_test_scaled = scaler.transform(X_test[['Trip_Distance_km', 'Per_Km_Rate', 'Trip_Duration_Minutes']])
```

```python
model = LinearRegression()
```

```python
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Mean Squared Error= {mse}")
print(f"Root Mean Squared Error={rmse}")
```

```python
r2 = r2_score(y_test, y_pred)  # Use actual & predicted values
print("R2 Score:", r2)
```

```python
y_predict=model.predict(X)
plt.scatter(y, y_predict, color='blue', label='Actual vs Predicted')

plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linestyle='dashed', label='Perfect Fit')

plt.xlabel('Actual Trip Price')
plt.ylabel('Predicted Trip Price')
plt.title('Multiple Regression: Actual vs. Predicted')
plt.legend()
plt.show()
```
